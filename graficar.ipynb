```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 3

list_x = []
list_y = []




for x in range (-100,101):
    list_x.append(x)
    list_y.append(f(x))
    

plt.plot(list_x, list_y, color='b')
plt.title('Grafica')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
