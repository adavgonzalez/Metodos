```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + 2 * x**2 + 10 * x - 20

def m(x):
    return 3 * x ** 2 + 4 * x + 10

def error(X_nuevo, X_anterior):
    return np.abs((X_nuevo - X_anterior) / X_nuevo)

a = 10
i = 0
iter = 10
Error = 100
list_i = []
list_xn = []
list_error = []
tolerancia = 0.01

while Error >= tolerancia:
    xn = a - (f(a)/ m(a))
    list_xn.append(xn)
    if xn < a:
        a = xn
    if i >= 1:
        Error = error(list_xn[-1], list_xn[-2])
    list_i.append(i)
    list_error.append(Error)
    print(f"En la iteracion {i} xn es igual a: {xn}")
    print(f"En la iteracion {i} el error es igual a: {Error}")
    print("")
    i += 1
print(f"la raiz es igual a: {xn}")

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(list_i, list_xn, color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteracin')
plt.ylabel('xn')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(list_i, list_error, color='r')
plt.title('Error Relativo')
plt.xlabel('Iteracin')
plt.ylabel('Error (%)')
plt.grid()

plt.tight_layout()
plt.show()
```
