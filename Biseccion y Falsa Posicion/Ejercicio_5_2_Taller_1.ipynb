```python
import numpy as np
import matplotlib.pyplot as plt

def funcion(x):
    return 4 * x ** 3 - 6 * x ** 2 + 7 * x - 2.3

def error(X_nuevo, X_anterior):
    return abs((X_nuevo - X_anterior) / X_nuevo) * 100

a = 0
b = 1

tolerancia = 1E-1

i = 0
list_xn = []
list_i = []
list_error = []
Error = 100

fa = funcion(a)
fb = funcion(b)

while Error >= tolerancia:
    if fa * fb == 0:
        if fa == 0:
            print(f"La raíz es {a}")
        elif fb == 0:
            print(f"La raíz es {b}")
        break

    elif fa * fb < 0:
        xn = (a + b) / 2
        fxn = funcion(xn)
        list_xn.append(xn)
        list_i.append(i)

        if i >= 1:
            X_nuevo = list_xn[i]
            X_anterior = list_xn[i - 1]
            Error = error(X_nuevo, X_anterior)
            list_error.append(Error)
            print(f"Error: {Error}")
        else:
            list_error.append(Error)

        if fxn == 0:
            print(f"La raíz es {xn}")
            break
        elif fa * fxn < 0:
            b = xn
            fb = fxn
        else:
            a = xn
            fa = fxn

        i += 1
        print(f"En la iteración {i}, xn es {xn}")
    else:
        print(f"No hay raíz en el intervalo [{a}, {b}]")
        break

print(f"La raíz es igual a {xn}")

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(list_i, list_xn, marker='o', color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteración')
plt.ylabel('xn')

plt.subplot(2, 1, 2)
plt.plot(list_i, list_error, marker='o', color='r')
plt.title('Error Relativo')
plt.xlabel('Iteración')
plt.ylabel('Error (%)')

plt.tight_layout()
plt.show()

```
