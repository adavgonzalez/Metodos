```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -1 + 5.5*x - 4*x**2 + 0.5*x**3 

def m(x): #derivada
    return 5.5 - 8*x + 1.5*x**2

def error(X_nuevo, X_anterior):
    return np.abs((X_nuevo - X_anterior) / X_nuevo)*100

a = 5 #valor inicial
i = 0
iter = 3
list_i = []
list_xn = []
list_error = [100]
tolerancia = 0.01

while list_error[-1] >= tolerancia:
    xn = a - (f(a) / m(a))
    list_xn.append(xn)
    
    list_i.append(i)
    if i >= 1:
        list_error.append(error(list_xn[-1], list_xn[-2]))
    print(f"En la iteracion {i} xn es igual a: {xn}")
    print(f"En la iteracion {i} el error es igual a: {list_error[-1]}%\n")
    print("------------------------------------------------------\n")
    a = xn  
    i += 1

print(f"la raiz es igual a: {xn}")

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(list_i, list_xn, color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteración')
plt.ylabel('xn')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(list_i, list_error, color='r')
plt.title('Error Relativo')
plt.xlabel('Iteración')
plt.ylabel('Error (%)')
plt.grid()

plt.tight_layout()
plt.show()
```
