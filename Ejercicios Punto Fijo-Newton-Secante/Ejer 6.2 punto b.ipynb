Punto Fijo 6.2 3 iteraciones X_{0} = 3

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

def error(X_nuevo, X_anterior):
    return abs((X_nuevo - X_anterior) / X_nuevo) * 100

def g(x):
    return (-2*(x**3) + 11.7*(x**2) + 5)/17.7

tolerancia = 0.01
iter = 10
x_0 = 3
    
list_error = [100]    
list_x = []
list_i = []
    
for i in range(iter):
    x = g(x_0)
        
    if(error(x , x_0) < tolerancia):
        print(f"En la iteracion {i + 1} se encuentra el punto fijo con un error de {error(x, x_0)}%")
            
        
        
    if i >= 1:
        list_error.append(error(x, x_0))
            
    list_x.append(x)
    list_i.append(i + 1)
    print(f"En la iteracion {i + 1} x = {x} con un error de {error(x, x_0)}%")    
    x_0 = x
        

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(list_i, list_x, marker='o', color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteración')
plt.ylabel('xn')

plt.subplot(2, 1, 2)
plt.plot(list_i, list_error, marker='o', color='r')
plt.title('Error Relativo')
plt.xlabel('Iteración')
plt.ylabel('Error (%)')

plt.tight_layout()
plt.show()
```
