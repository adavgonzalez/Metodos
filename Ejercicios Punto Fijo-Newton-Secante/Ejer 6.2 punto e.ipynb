Secante tres iteraciones rango [3,4] 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*(x**3) - 11.7*(x**2) + 17.7*(x) - 5

def m(xi , xf):
    return (f(xf) - f(xi))/(xf - xi)

def error(X_nuevo, X_anterior):
    return np.abs((X_nuevo - X_anterior) / X_nuevo)

# Valores iniciales
a = 3
b = a + 0.01  
i = 0
iter = 3
Error = 100
list_i = []
list_xn = []
list_error = []
tolerancia = 0.01

while iter >= i:
    xn = a - f(a)*( 1/ m(a, b))
    list_xn.append(xn)
    if xn < a:
        b = a
        a = xn
    else:
        a = b
        b = xn
    if i >= 1:
        Error = error(list_xn[-1], list_xn[-2])
    list_i.append(i)
    list_error.append(Error)
    print(f"En la iteracion {i} xn es igual a: {xn}")
    print(f"En la iteracion {i} el error es igual a: {Error}%")
    print("")
    i += 1


plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(list_i, list_xn, color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteracin')
plt.ylabel('xn')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(list_i, list_error, color='r')
plt.title('Error Relativo')
plt.xlabel('Iteracin')
plt.ylabel('Error (%)')
plt.grid()

plt.tight_layout()
plt.show()
```
