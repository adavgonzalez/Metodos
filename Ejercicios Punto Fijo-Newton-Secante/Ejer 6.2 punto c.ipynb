Newton-raphson 3 iteraciones x inicial = 3

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*(x**3) - 11.7*(x**2) + 17.7*(x) - 5

def m(x): #derivada
    return 6*(x**2) - 23.4*x + 17.7

def error(X_nuevo, X_anterior):
    return np.abs((X_nuevo - X_anterior) / X_nuevo)*100

a = 3 #valor inicial
i = 0
iter = 3
list_i = []
list_xn = []
list_error = []
tolerancia = 0.01

while iter >= i:
    xn = a - (f(a) / m(a))
    list_xn.append(xn)
    Error = error(list_xn[-1], list_xn[-2]) if i >= 1 else 100
    list_i.append(i)
    list_error.append(Error)
    print(f"En la iteracion {i} xn es igual a: {xn}")
    print(f"En la iteracion {i} el error es igual a: {Error}%\n")
    print("------------------------------------------------------\n")
    a = xn  
    i += 1

print(f"la raiz es igual a: {xn}")

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(list_i, list_xn, color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteración')
plt.ylabel('xn')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(list_i, list_error, color='r')
plt.title('Error Relativo')
plt.xlabel('Iteración')
plt.ylabel('Error (%)')
plt.grid()

plt.tight_layout()
plt.show()
```
