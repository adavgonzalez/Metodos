```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
  return 2 * x ** 3 - 11.7 * x ** 2 + 18.7 * x - 5

def error(X_actual, X_antiguo):
  return np.abs((X_actual - X_antiguo)/X_actual)

x_ini = 3
N_iter = 3
i = 0
x = x_ini
list_x = []
list_i = []
list_error = []

tolerancia = 1E-2

while(N_iter > i):
  
  x = g(x)
  list_x.append(x)
  
  if i == 0:
    
    list_error.append(100)
  
  elif i >= 1:
    
    list_error.append(error(list_x[i], list_x[i - 1]))
  
  print(f"Error = {list_error[i]}")
  print(f"El valor de x evaluado en g(x) es = {x}")
  print("")
  i += 1
  list_i.append(i)


plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(list_i, list_x, marker='o', color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteración')
plt.ylabel('xn')

plt.subplot(2, 1, 2)
plt.plot(list_i, list_error, marker='o', color='r')
plt.title('Error Relativo')
plt.xlabel('Iteración')
plt.ylabel('Error (%)')

plt.tight_layout()
plt.show()
```
