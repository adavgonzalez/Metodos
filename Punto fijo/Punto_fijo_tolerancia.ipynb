```python
from math import e
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
  return np.exp(-x)

def error(X_actual, X_antiguo):
  return abs((X_actual - X_antiguo)/X_actual)*100
```

```python
x_ini = 0
N_iter = 10
i=0
x = x_ini
list_x = []
list_i = []
list_error = [100]
tolerancia = 1E-2
```

```python
while(list_error[-1] >= tolerancia):
  x = g(x)
  list_x.append(x)
  list_i.append(i)
  if i == 0:
    list_error.append(100)
  elif i > 1:
    list_error.append(error(list_x[-1], list_x[-2]))
  print(f"Error = {list_error[-1]}")
  print("")
  print(f"El valor de x evaluado en g(x) es = {x}")
  i += 1


plt.plot(list_i, list_error)
plt.show()

plt.plot(list_i , list_x)
plt.show()
```
